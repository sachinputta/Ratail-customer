```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('Customer_Buying_Patterns.csv')
```

```python
df.head()
```

```python
df['CategoryName'] = df['CategoryName'].str.strip()
df.dropna(axis=0, subset=['BillNo'], inplace=True)
df['BillNo'] = df['BillNo'].astype('str')
df = df[~df['BillNo'].str.contains('C')]
```

```python
df.head()
```

```python
df["ProductName"]=df["ProductName"].astype("category").cat.codes
```

```python
df.dtypes
```

```python
tr={}
```

```python
data=df[["BillNo","ProductName"]]
    
```

```python
len(data["BillNo"].unique())
```

```python
data=data.values
```

```python
for i in data:
    if  i[0]  in tr.keys():
        tr[i[0]].append(i[1])
    else :
        tr[i[0]]=[i[1]]
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(tr).transform(tr)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
```

```python
tr
```

```python
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

res = association_rules(frequent_itemsets, support_only=True, min_threshold=0.1)
res.head(5)
```
